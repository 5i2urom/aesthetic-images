```python
tags_names = ["Tag ID", "Name"]

# Читаем файл построчно
with open("/kaggle/input/ava-aesthetic-visual-assessment/AVA_Files/tags.txt", "r") as file:
    rows = []
    for line in file:
        # Разделяем строку по первому пробелу (чтобы сохранить текст с пробелами в "Name")
        parts = line.strip().split(" ", 1)
        if len(parts) == 2:  # Убедимся, что есть и ID, и Name
            rows.append(parts)

# Преобразуем данные в DataFrame
tags = pd.DataFrame(rows, columns=tags_names)
```

```python
# Вычисляем общее количество голосов
ratings = ava.iloc[:, 2:12].values  # Колонки с "Rating 1" по "Rating 10"
total_votes = np.sum(ratings, axis=1)

# Вычисляем среднюю оценку и дисперсию
ava['MeanRating'] = np.sum(ratings * np.arange(1, 11), axis=1) / total_votes
ava['Variance'] = np.sum(ratings * (np.arange(1, 11) - ava['MeanRating'].values[:, None])**2, axis=1) / total_votes
```

## Определение нормальности

```python
from scipy.stats import shapiro

# Функция для проверки нормальности распределений
def check_normality(row):
    ratings = row[["Rating 1", "Rating 2", "Rating 3", "Rating 4", 
                   "Rating 5", "Rating 6", "Rating 7", "Rating 8", 
                   "Rating 9", "Rating 10"]].values
    # Нормализуем распределение, чтобы получить вероятности
    total_votes = ratings.sum()
    if total_votes == 0:
        return "Unknown"  # Если нет голосов, классифицируем как "неизвестно"
    probabilities = ratings / total_votes
    stat, p = shapiro(probabilities)
    return "Normal" if p > 0.05 else "Not Normal"

# Применяем функцию ко всему DataFrame
ava['Distribution Type'] = ava.apply(check_normality, axis=1)

# Подсчёт количества нормальных и ненормальных распределений
distribution_counts = ava['Distribution Type'].value_counts()

print(distribution_counts)
```

## Фильтрация "неоднозначных" оценок

```python
δ = 0.5
filtered_ava = ava[(ava['MeanRating'] < (5 - δ)) | (ava['MeanRating'] > (5 + δ))]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(ava['MeanRating'], bins=20, kde=True, color='blue', label='До фильтрации')
sns.histplot(filtered_ava['MeanRating'], bins=20, kde=True, color='orange', label='После фильтрации')
plt.title('Распределение средних оценок до и после фильтрации')
plt.xlabel('Средняя оценка')
plt.ylabel('Количество изображений')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ava['Variance'], bins=20, kde=True, color='green', label='До фильтрации')
sns.histplot(filtered_ava['Variance'], bins=20, kde=True, color='red', label='После фильтрации')
plt.title('Распределение дисперсий до и после фильтрации')
plt.xlabel('Дисперсия')
plt.ylabel('Количество изображений')
plt.legend()
plt.show()
```
